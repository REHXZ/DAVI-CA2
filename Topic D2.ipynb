```python
sns.set_theme(style="whitegrid")

df= pd.read_csv("DAVI Dataset/movie_scores.csv", index_col=0)
df = pd.melt(df, id_vars=["MovieTitle"],var_name='Type',
    value_vars=['Tomatometer','AudienceScore'],
    value_name='Rating')

plt.figure(figsize=(10, 6))

barplot = sns.barplot(x="MovieTitle", y="Rating",hue="Type",
    data=df, hue_order=['AudienceScore','Tomatometer'])

plt.xticks(rotation=45)
plt.legend(loc='upper right')
plt.show()
```

```python
data= pd.read_csv ("scores.csv")
data
```

```python
A = list(data['group_a'])
B = list(data['group_b'])
C = list(data['group_c'])
D = list(data['group_d'])
```

```python
new_data = {
    'Group A': A,
    'Group B': B,
    'Group C': C,
    'Group D': D
}

new_dataframe = pd.DataFrame(new_data)
melted_data = new_dataframe.melt(var_name='Group', value_name='IQ Score')

# Step 5: Customize the plot
violin_plot = sns.violinplot(x='Group', y='IQ Score', data=melted_data)
violin_plot.set_title('IQ Scores by Test Group')
sns.despine(top=True, right=True, left=True)

# Show the plot
plt.show()
```
