```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.set_style("whitegrid")
sns.barplot(x=["A", "B", "C"], y=[1, 3, 2])

plt.show()
```

```python
%matplotlib inline
sns.set_style("whitegrid")

plt.figure()
x1 = [10,20,5,40,8]
x2=[30,40,10,20,30]

with sns.axes_style('ticks'):
    plt.plot(x1, label = 'Group A')
    plt.plot(x2, label = 'Group B')
plt.legend()
plt.show()
```

```python
%matplotlib inline
sns.set_style("whitegrid")

plt.figure()
x1 = [10,20,5,40,8]
x2=[30,40,10,20,30]

with sns.axes_style('ticks'):
    plt.plot(x1, label = 'Group A')
    plt.plot(x2, label = 'Group B')

sns.despine(fig=None, ax=None, top=True, right=True, left=True, bottom=True, offset=None, 
trim=False)
# plt.legend()
plt.show()
```

```python
tips=sns.load_dataset('tips')
sns.boxplot(x='day', y='total_bill', data=tips)
#remove top and right axes spines
sns.despine(fig=None, top=True, right=True)
plt.show()
```

```python
df = pd.read_csv('scores.csv')
df.rename(columns={'group_a': 'Group A', 'group_b': 'Group B', 'group_c': 'Group C', 'group_d':'Group D'}, inplace=True)
sns.boxplot( data=df)
sns.despine(fig=None, top=True, right=True)
plt.xlabel('Groups')
plt.ylabel('IQ Score')
plt.show()
```
